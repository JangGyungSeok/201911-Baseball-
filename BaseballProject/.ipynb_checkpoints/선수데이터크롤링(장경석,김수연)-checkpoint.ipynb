```python
# =================================== 장경석 , 김수연 작성 파일 =======================================

#크롤링을 위한 라이브러리 추가

import time
import requests
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import urllib.request
from PIL import Image
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
```

```python
# MLB 선수 데이터를 담은 사이트와 크롤링을 위한 driver 선언
url1='https://www.fangraphs.com/'
driver1 = webdriver.Chrome('chromedriver.exe')
# driver2 = webdriver.Chrome(executable_path='C:/ai/workspace/python/chromedriver.exe')
# driver1.get(
# driver2.get()
```

```python

# 이건 추후에 웹에서 입력받는 것으로 수정
pitcher_name='Hyun-Jin Ryu'

# 선수명 검색바에 선수명입력 및 search 버튼 클릭
driver = webdriver.Chrome('chromedriver.exe')
driver.get(url1)
time.sleep(3)
driver.find_element_by_class_name('search-query').send_keys(pitcher_name)
time.sleep(1)
driver.find_element_by_class_name('search-query').send_keys(Keys.ENTER)
time.sleep(1)

```

```python
# 선수명으로 검색된 페이지의 소스를 받는다.
html=driver.page_source
soup = BeautifulSoup(html,'html.parser')
```

```python
# 선수 정보 테이블을 담고 있는 객체
contents=soup.find_all('div',{'style':'white-space:nowrap;'})
contents
```

```python
# re를 통해 테이블에 숨겨져 있는 쓸모없는 값을 필터링해줌
# 쓸모없는 객체의 속성이 담고있는 특정 문자열로 분류해줌
import re
comp=re.compile('grid_multi')
comp2=re.compile('[(]')
comp3=re.compile('grid_postseason')
comp4=re.compile('rgRow ')
comp5=re.compile('rgAltRow ')

```

```python
# 페이지 소스에서 실질적으로 사용할 값만 사용하고 저장하기 위한 dic 선언
data_list={
    'team':[],'year':[],'BABIP':[],'ERA':[],'WAR':[],'TBF':[],'H':[],'HR':[],'BB':[],'HBP':[],'SO':[],'WHIP':[],'FIP':[],
    'LD':[],'GB':[],'FB':[],'IFFB':[],'Swing':[],'SWStr':[]
          }


# 각 테이블별로 원하는 정보를 추출해옴
# for 문 내부의 알고리즘설명은 생략

bodys=contents[0].select('tbody')
for body in bodys:
    trs=body.select('tr')
    for tr in trs[:-2]:
        tds=tr.select('td')
        season=tds[0].text
        team=tds[1].text
        
        year=tds[0].text
        BABIP=tds[11].text
        ERA=tds[15].text
        WAR=tds[18].text
        if comp.search(str(tr))==None and comp3.search(str(tr))==None and season!='2020' and season!='Total':
            if comp2.search(team)==None:
                data_list['year'].append(year)
                data_list['BABIP'].append(BABIP)
                data_list['ERA'].append(ERA)
                data_list['WAR'].append(WAR)
                data_list['team'].append(team)

# #==========================================================================================
bodys=contents[1].select('tbody')
for body in bodys:
    trs=body.select('tr')
    for tr in trs[:-1]:
        tds=tr.select('td')
        team=tds[1].text
        season=tds[0].text

        TBF=tds[13].text
        H=  tds[14].text
        HR= tds[17].text
        BB= tds[18].text
        HBP=tds[20].text
        SO= tds[23].text
        if comp.search(str(tr))==None and comp3.search(str(tr))==None and season!='2020' and season!='Total':
            if comp2.search(team)==None:
                data_list['TBF'].append(TBF)
                data_list['H'].append(H)
                data_list['HR'].append(HR)
                data_list['BB'].append(BB)
                data_list['HBP'].append(HBP)
                data_list['SO'].append(SO)
                print(season)

                print(team)
# #==========================================================================================
bodys=contents[2].select('tbody')
for body in bodys:
    trs=body.select('tr')
    for tr in trs[:-1]:
        tds=tr.select('td')
        team=tds[1].text
        season=tds[0].text

        WHIP= tds[10].text
        FIP= tds[15].text
        if comp.search(str(tr))==None and comp4.search(str(tr))==None and comp5.search(str(tr))==None and season!='2020' and season!='Total':
            if comp2.search(team)==None:
                data_list['WHIP'].append(WHIP)
                data_list['FIP'].append(FIP)
# #==========================================================================================
bodys=contents[3].select('tbody')
for body in bodys:
    trs=body.select('tr')
    for tr in trs[:-1]:
        tds=tr.select('td')
        season=tds[0].text
        team=tds[1].text
        
        LD=tds[3].text
        GB=tds[4].text
        FB=tds[5].text
        IFFB=tds[6].text
        if comp.search(str(tr))==None and comp4.search(str(tr))==None and comp5.search(str(tr))==None and season!='2020' and season!='Total':
            if comp2.search(team)==None:
                data_list['LD'].append(LD)
                data_list['GB'].append(GB)
                data_list['FB'].append(FB)
                data_list['IFFB'].append(IFFB)
                
                

# #==========================================================================================
bodys=contents[11].select('tbody')
for body in bodys:
    trs=body.select('tr')
    for tr in trs[:-1]:
        tds=tr.select('td')
        season=tds[0].text
        team=tds[1].text
        
        Swing= tds[4].text
        SWStr= tds[10].text
        if comp.search(str(tr))==None and season!='2020' and season!='Total':
            if comp2.search(team)==None:
                data_list['Swing'].append(Swing)
                data_list['SWStr'].append(SWStr)
                #print(season)
# 데이터를 받아와 저장한 dic를 출력
data_list
```
